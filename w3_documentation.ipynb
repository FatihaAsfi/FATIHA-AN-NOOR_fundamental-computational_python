# Week 3 Tutorial 3

## 1. Components 

### Input 
- Quiz 1 mark 
- Quiz 2 mark 
- Quiz 3 mark 
- Continue choice 

### Process 
1. Accept three quiz marks.
2. Calculate average.
3. Check whether average is 50 or above.
4. Display Pass or Fail.
5. Ask user whether to continue. 

### Output 
- Average mark
- Pass/Fail status 
- Program Ended message

## 2. Psuedocode 

START 

Set choice = "Y"

WHILE choice == "Y"

    Input quiz_1
    Input quiz_2
    Input quiz_3

    average = (quiz_1 + quiz_2 + quiz_3) / 3

    Display average 

    IF average >= 50
         Display PASS
    ELSE 
         Display FAIL
    ENDIF

    Input choice 

    ENDWHILE 

    Display "Program Ended"

    STOP

## 2.2 Algorithm Structure

### Sequence 
- Input quiz marks 
- Calculate Average 
- Display Average 

### Selection 
- IF average >= 50 THEN PASS
- ELSE FAIL

### Iteration 
- WHILE choice == "Y"
